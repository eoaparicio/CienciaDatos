# Definición de features

```python
#Obtenemos los features como un tensor de orden 2
x = df.values
x.shape
```

# Calibración de K

```python
#Importamos el objeto para clusterizar
from sklearn.cluster import KMeans
```

```python
#Valores de K que se probarán
k_max = 15
```

```python
#Inercia de cada K -- La inercia mejor es la más pequeña.
inercia = []
```

```python
#Probamos cada K
for k in range(1, k_max):
  #Definimos un modelo
  model = KMeans(
      n_clusters=k, 
      init='random',
      n_init=10, 
      max_iter=300, 
      tol=1e-04, 
      random_state=0
  )

  #Entrenamos, no queremos asociar, por eso no se le pone predict.
  model.fit(x)

  #Almacenamos la inercia
  inercia.append(model.inertia_)
```

# Gráfico de inercia

```python
#Importamos la librería para graficar
import matplotlib.pyplot as plt
```

```python
#Grafimos los k y sus inercias
plt.plot(
    range(1, k_max), 
    inercia, 
    marker='o'
)

#Agregamos las etiquetas
plt.xlabel('K')
plt.ylabel('Inercia')

#Mostramos el grafico
plt.show()
```
