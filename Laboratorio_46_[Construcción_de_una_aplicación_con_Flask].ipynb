# Creación de una Web Flask

```python
#Importamos la librería Flask
import flask
```

```python
#Importamos el objeto Flask
from flask import Flask
```

```python
#Importamos la librería para ver el servidor Flask desde nuestro navegador
from flask_ngrok import run_with_ngrok
```

```python
#Creamos una aplicación web con Flask
#"__name__" es el nombre de nuestro proyecto de Google Colab
app = Flask(__name__)
```

```python
#Indicamos que queremos ejecutar nuestra aplicación web con NGrok para poder ver la web
run_with_ngrok(app)
```

```python
#Creamos la función que define el contenido de la raíz de la web (/)
#De manera estándar a la raíz se le conoce como "home"
@app.route("/")
def home():
    return "<h1>Servidor Flask ejecutandose...</h1>"
```

```python
#Creamos otra web en el servidor
@app.route("/saluda")
def saluda():
    #Capturamos el parametro que recibe la web
    nombre = flask.request.args.get('nombre')

    #Mostramos la web con el parametro
    return "<h1>Hola a "+nombre+"<h1>"
```

```python
#Ejecutamos la aplicación
#Para abrir la web deberemos dar clic en el enlace que aparezca
#Para probar el método "saluda" deberemos probar la ruta: saluda?nombre=Alonso
#Para detener el servidor deberemos detener la ejecución de esta sección del código
app.run()
```
